```python
import os
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
df_train = pd.DataFrame(columns=['text', 'rating', 'status'])
df_test = pd.DataFrame(columns=['text', 'rating', 'status'])
```

```python
os.chdir('/content/aclImdb/train')
for status in ['neg', 'pos']:
  for curr_file in os.listdir(f'./{status}'):
    with open(f'./{status}/{curr_file}', 'r') as f:
      text = f.read()
      rating = curr_file.split('.')[0].split('_')[1]
    df_train = df_train.append({'text': text, 'rating': rating, 'status': status}, ignore_index=True)
```

```python
df_train
```

```python
os.chdir('/content/aclImdb/test')
for status in ['neg', 'pos']:
  for curr_file in os.listdir(f'./{status}'):
    with open(f'./{status}/{curr_file}', 'r') as f:
      text = f.read()
      rating = curr_file.split('.')[0].split('_')[1]
    df_test = df_test.append({'text': text, 'rating': rating, 'status': status}, ignore_index=True)
```

```python
df_test
```

```python
df_train['status'] = df_train['status'].replace({'neg': 0, 'pos': 1})
df_test['status'] = df_test['status'].replace({'neg': 0, 'pos': 1})
```

```python
df_train.to_csv('/content/drive/MyDrive/greenatom/train.csv')
df_test.to_csv('/content/drive/MyDrive/greenatom/test.csv')
```
